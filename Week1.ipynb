```python
import numpy as np # is an array manipulation, used ofr linear algebra, Fourier transform, and random number capabilities.
import pandas as pd # is a library for data manipulation and data analysis.
import cv2 as cv # is a library for computer vision tasks.
from google.colab.patches import cv2_imshow # for image display
from skimage import io # skimage is a library which supports image processing application on phyton.
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt # Matplotlib is a library which generates figures and provides graphical user interface toolkit.
```

Langkah 2

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)                           #read image
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)  #resize image to half size
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)   #convert color to RGB
    final_frame = cv.hconcat([image, image_2])       #concatenate image
    cv2_imshow(final_frame)                          #show image
    print('\n')
```

```python
urls = "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg"
image = io.imread(urls)
image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
tinggi, lebar, _ = image_2.shape

# Membuat garis vertikal ditengah image
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for x in range(tinggi):
  image_3[x, int((lebar)/2)] = [255,255,255]

# Membuat garis menyilang diagonal
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for y in range(lebar):
  image_4[int(tinggi / lebar * y), y] = [255,255,255] # garis diagonal dari kiri atas ke kanan bawah
  image_4[int(tinggi / lebar * y), -y] = [255,255,255] # garis diagonal dari kanan atas ke kiri bawah

# Menyusun gambar
vertikal = cv.hconcat((image_2, image_3))
menyilang = cv.hconcat((image_2, image_4))

# Menampilkan gambar
cv2_imshow(vertikal)
print('\n')
cv2_imshow(menyilang)
```

```python
urls = "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg"
image = io.imread(urls)
image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
tinggi, lebar, _ = image_2.shape

# Membuat garis horisontal berwarna putih dibagian tengah gambar dengan panjang tertentu
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
garis_panjang = 100  # Panjang garis horizontal
mid_y = tinggi // 2  # Posisi vertikal tengah

# Menghitung posisi awal dan akhir garis agar berada di tengah
start_x = max((lebar - garis_panjang - 60), 0)
end_x = min(start_x + garis_panjang, lebar)

# Menggambar garis horizontal putih
image_5[mid_y, start_x:end_x] = [255, 255, 255]

# Menampilkan gambar
print('Image with Horizontal Line')
final_frame = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame)
```

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg"]
image = io.imread(urls[0])
image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
tinggi, lebar, _ = image_2.shape

# Koordinat dan ukuran kotak
x_start = 120  # Koordinat x dari sudut kiri atas kotak
y_start = 45  # Koordinat y dari sudut kiri atas kotak
width = 90  # Lebar kotak
height = 40  # Tinggi kotak

# Membuat gambar salinan untuk modifikasi
image_with_box = image_2.copy()

# Menentukan koordinat akhir kotak
x_end = min(x_start + width, lebar)
y_end = min(y_start + height, tinggi)

# Menggambar kotak putih
image_with_box[y_start:y_end, x_start:x_end] = [255, 255, 255]

# Menampilkan gambar
print('Image with White Box')
final_frame = cv.hconcat((image_2, image_with_box))
cv2_imshow(final_frame)
```
