```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

# Membaca gambar
img = cv.imread('/content/drive/MyDrive/Kuliah polinema/sem 5/PCVK/Images/kucink.jpg')

# Mengonversi dari BGR ke RGB agar sesuai dengan tampilan Matplotlib
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Membuat salinan untuk masing-masing channel (RB dan GB)
img_rb = img_rgb.copy()  # Red-Blue
img_gb = img_rgb.copy()  # Green-Blue

# Menonaktifkan channel Green untuk RB (index 1)
img_rb[:, :, 1] = 0  # Channel Green di-nol-kan

# Menonaktifkan channel Red untuk GB (index 0)
img_gb[:, :, 0] = 0  # Channel Red di-nol-kan

# Menampilkan gambar
plt.figure(figsize=(10, 5))

# Gambar Red-Blue (RB)
plt.subplot(1, 2, 1) # Membuat subplot 1 dari grid 1 baris, 2 kolom (subplot di sebelah kiri)
plt.title("Red-Blue (RB) Channel")
plt.imshow(img_rb)

# Gambar Green-Blue (GB)
plt.subplot(1, 2, 2)
plt.title("Green-Blue (GB) Channel")
plt.imshow(img_gb)

plt.show()
```

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

# Membaca gambar
img = cv.imread('/content/drive/MyDrive/Kuliah polinema/sem 5/PCVK/Images/kucink.jpg')

# Mengonversi dari BGR ke RGB agar sesuai dengan tampilan Matplotlib
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Melakukan cropping (baris 20-115, kolom 25-120)
cropped_img = img_rgb[20:116, 25:121]

# Menampilkan gambar yang telah di-crop
plt.imshow(cropped_img)
plt.title('Cropped Image (Rows 20-115, Columns 25-120)')
plt.axis('on')  # Menampilkan axis untuk referensi posisi
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Membaca gambar
img = cv.imread('/content/drive/MyDrive/Kuliah polinema/sem 5/PCVK/Images/kucink.jpg')

# Mengonversi dari BGR ke RGB agar sesuai dengan tampilan Matplotlib
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Menyimpan hanya channel Red
red_channel = img_rgb[:, :, 0]  # Channel Red ada di index 0

# Melakukan cropping (baris 5-30, semua kolom)
cropped_red = red_channel[5:31, :]  # baris 5 hingga 30 (index 5 sampai 30), semua kolom

# Menampilkan gambar channel Red yang telah di-crop
plt.imshow(cropped_red, cmap='gray')
plt.title('Cropped Red Channel (Rows 5-30)')
plt.axis('on')  # Menampilkan axis untuk referensi posisi
plt.show()
```
